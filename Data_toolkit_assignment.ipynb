```python
# Explain the following termsg
1. The difference in np.array, np.asarray and np.asanyarray.
2. The difference between Deep copy and shallow copy.


1. Difference between np.array, np.asarray, and np.asanyarray

np.array:
This function creates a new ndarray object from any given input, even if the input is already an array-like object (e.g., a list, tuple, or even an existing NumPy array).

Key Point: np.array will always copy the input data by default (unless you explicitly specify copy=False).


arr1 = np.array([1, 2, 3])  # Creates a new array
arr2 = np.array(arr1)  # Copies arr1 into a new array


np.asarray:
This function is similar to np.array, but it won't copy the data if the input is already an ndarray. It only converts input to an array if it is not already one.

Key Point: It does not make a copy if the input is already a NumPy array.

arr1 = np.array([1, 2, 3])
arr2 = np.asarray(arr1)  # Will not copy the data, just reference arr1


np.asanyarray:
This function is even more relaxed than np.asarray. It behaves like np.asarray, but if the input is a subclass of ndarray (like matrix), it will return the same subclass instead of forcing it into a base ndarray.

Key Point: It allows the input to remain a subclass of ndarray if applicable.

arr = np.matrix([[1, 2], [3, 4]])
arr2 = np.asanyarray(arr)  # Retains the matrix type



2. Difference between Deep Copy and Shallow Copy

Shallow Copy:
A shallow copy creates a new object but does not copy the nested objects. Instead, it just copies the references to the nested objects.
This means that changes made to mutable nested objects in the shallow copy will reflect in the original object because both will refer to the same objects.


import copy
list1 = [[1, 2], [3, 4]]
list2 = copy.copy(list1)  # Shallow copy
list2[0][0] = 99
print(list1)  # [[99, 2], [3, 4]] (reflects change in the original list)


Deep Copy:
A deep copy creates a completely independent copy of both the object and all of its nested objects. Changes in the deep copy do not affect the original object, as all objects are recursively copied.


list1 = [[1, 2], [3, 4]]
list2 = copy.deepcopy(list1)  # Deep copy
list2[0][0] = 99
print(list1)  # [[1, 2], [3, 4]] (no changes in the original list)

```

```python
# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)

# Create a DataFrame to hold the data
data = pd.DataFrame({'X': x, 'Y': y})

# Define a function to categorize points into quadrants
def get_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant 4'
    else:
        return 'Origin/Axis'

# Apply the function to determine the quadrant for each point
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', s=100)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]  # 10 categories
values = np.random.randint(10, 100, size=10)  # Random values for each category

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Define colors based on values
colors = ['#%02x%02x%02x' % (int(255 - (value * 2.55)), 0, int(value * 2.55) ) for value in values]

# Create a new plot with title and axis labels
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

# Add bars to the plot
p.vbar(x='Category', top='Value', source=source, width=0.9, color=colors)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Show grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
output_notebook()  # Use this if you're running in a Jupyter Notebook
show(p)
```
